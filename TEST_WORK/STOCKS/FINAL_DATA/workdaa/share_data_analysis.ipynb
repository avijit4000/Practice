```python
def omit_personal_titles(name):
    # List of personal name titles to be omitted
    titles = ["Mr.", "Ms.", "Dr.", "Miss", "Sir"]

    # Split the name into words
    words = name.split()

    # Filter out words that are not personal name titles
    filtered_words = [word for word in words if word not in titles]

    # Join the filtered words back into a single string
    cleaned_name = ' '.join(filtered_words)

    return cleaned_name

# Example usage
if __name__ == "__main__":
    for created_by in cmpany_name[8]:
        cleaned_name = omit_personal_titles(created_by)
        print(cleaned_name)
```
