```python
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
```

```python
df_tm_d=pd.read_csv('df_tm_d.csv')
df_bel_d=pd.read_csv('df_bel_d.csv')
df_nmdc_d=pd.read_csv('df_nmdc_d.csv')
df_hdfc_d=pd.read_csv('df_hdfc_d.csv')
df_ongc_d=pd.read_csv('df_ongc_d.csv')
```

```python
df_tm_d=df_tm_d.rename(columns={'Close':'tm_Clo',
                                'Volume':'tm_Vol'})
```

```python
df_bel_d=df_bel_d.rename(columns={'Close':'bel_Clo',
                                'Volume':'bel_Vol'})
df_nmdc_d=df_nmdc_d.rename(columns={'Close':'nmdc_Clo',
                                'Volume':'nmdc_Vol'})
df_hdfc_d=df_hdfc_d.rename(columns={'Close':'hdfc_Clo',
                                'Volume':'hdfc_Vol'})
df_ongc_d=df_ongc_d.rename(columns={'Close':'ongc_Clo',
                                'Volume':'ongc_Vol'})
```

```python
result = pd.concat([df_tm_d["tm_Vol"],df_tm_d["tm_Clo"],df_bel_d["bel_Vol"],df_bel_d["bel_Clo"],df_nmdc_d["nmdc_Vol"],df_nmdc_d["nmdc_Clo"], df_ongc_d['ongc_Vol'],df_ongc_d['ongc_Clo'],df_hdfc_d["hdfc_Vol"],df_hdfc_d["hdfc_Clo"]], axis=1)
```

```python
result_corr_matrix = result.corr().round(2) #[-1 to +1]
```

```python
scaler=StandardScaler()
scaler.fit(result)
```

```python
scaled_data=scaler.transform(result)
```

```python
scaled_data
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```
