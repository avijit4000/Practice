```python
import pandas as pd

# Provided list data
data = [
    "1.1 HE im 19'' Rack 2,000 65,000 € 130,00 €",
    "01.03.2023 - 31.03.2023 (+ 0,00 €) 0%",
    "2.IP-Netz (159.100.29.0/24, 212.224.86.0/24, 212.224.88.0/24, 8,000 154,500 € 1.236,00 €",
    "Abrechnung für 01.03.2023 - 31.03.2023 (+ 0,00 €) 0%",
    "3.Remote Hands - Service Dienstleistung 1,000 45,000 € 45,00 €",
    "Ihr Auftrag im Ticket #FC-35780 - 24.02.2023 (+ 0,00 €) 0%",
    "4.Remote Hands - Service Dienstleistung 3,000 45,000 € 135,00 €",
    "Ihr Auftrag im Ticket #FC-35830 - 08.02.2023 (+ 0,00 €) 0%",
    "5.Remote Hands - Service Dienstleistung 1,000 45,000 € 45,00 €",
    "Ihr Auftrag im Ticket #FC-35975 - 08.02.2023 (+ 0,00 €) 0%",
    "6.Remote Hands - Service Dienstleistung 8,000 45,000 € 360,00 €",
    "Ihr Auftrag im Ticket #FC-35822 - 01.02.2023 (+ 0,00 €) 0%",
    "7.Remote Hands - Service Dienstleistung 3,000 45,000 € 135,00 €",
    "Ihr Auftrag im Ticket #FC-36167 - 23.02.2023 (+ 0,00 €) 0%"
]

# Parsing the list into structured data
parsed_data = []

for i in range(0, len(data), 2):
    description = data[i]
    date_and_cost = data[i + 1]

    # Split the description to extract numeric values and cost
    desc_parts = description.split()
    numeric_values = [part for part in desc_parts if ',' in part or part.isdigit()]
    description_text = " ".join(part for part in desc_parts if part not in numeric_values)

    # Extracting individual numeric values
    value_1 = numeric_values[-3].replace(',', '.')
    value_2 = numeric_values[-2].replace(',', '.').replace('€', '')
    value_3 = numeric_values[-1].replace(',', '.').replace('€', '')

    # Extract date range and additional costs
    date_range = date_and_cost.split(' (+')[0]
    additional_cost = date_and_cost.split(' (+')[1].replace(') 0%', '')

    # Append parsed data to the list
    parsed_data.append([description_text, str(value_1), str(value_2), str(value_3), date_range, additional_cost])

# Creating DataFrame
df = pd.DataFrame(parsed_data, columns=['Description', 'Value1', 'Value2', 'Value3', 'Date Range', 'Additional Cost'])

# # Display the DataFrame
# print(df)
```
